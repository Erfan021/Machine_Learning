#### Matplotlib workflow

```python
# 0. Import Matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [10,11,12,13]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5,5))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title='Sample plot',
      xlabel='x-axis',
      ylabel='y-axis')

# 5. Save & show
fig.savefig('images/sample-plot.png')
```

```python
# Make plot form dictionary
# Bar plot
butter_prices = {'Almond butter': 10,
                'Peanut Butter': 8,
                'Cashew Butter': 12}
fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values())
ax.set(title='Butter store',
      ylabel='Prices')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()))
ax.set(title='Butter store',
      xlabel='Prices')
plt.show()
```

```python
# Histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Multiple plots at a time
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
```

```python
# Subplots option:1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
# plot to each different axes
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(butter_prices.keys(), butter_prices.values());
ax4.hist(np.random.randn(100));
```

```python
# Subplots option:2
fig, ax  = plt.subplots(nrows=2,
                        ncols=2,
                        figsize=(10,5))
# plot to each different index axes
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(butter_prices.keys(), butter_prices.values());
ax[1,1].hist(np.random.randn(100));
```

```python
hear_diease = pd.read_csv("D:\\TUTORIALS\\Exercise Files - Complete Machine Learning and Data Science Zero to Mastery\\4- Matplotlib/heartdiease.csv")
```

```python
over_fifty = hear_diease[hear_diease['age'] > 50]
over_fifty.head()
```

```python
# PyPlot Method
over_fifty.plot(kind='scatter', 
                x='age', y='chol',
                c='target');
```

```python
# Object-Oriented Method with PyPlot method
fig, ax = plt.subplots(figsize=(10,6))
over_fifty.plot(kind='scatter',
               x='age', y='chol',
               c='target',
               ax=ax);
#ax.set_xlim((45,100))
```

```python
# Object-oriented Method
fig, ax = plt.subplots(figsize=(10,6))

# data plotting
scatter = ax.scatter(x=over_fifty['age'],
                    y=over_fifty['chol'],
                    c=over_fifty['target'])

# setting labels and title
ax.set(title='Heart diease and cholestrol levels',
       xlabel='Age',
      ylabel='Cholestrol')

# adding legend
ax.legend(*scatter.legend_elements(), title='Target')

# adding horizontal line
ax.axhline(over_fifty['chol'].mean(),
          linestyle='--')

plt.show()
```

```python
over_fifty.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10),
                              nrows=2,
                              ncols=1,
                              sharex=True)

# adding data to ax0
scatter = ax0.scatter(x=over_fifty['age'],
                     y=over_fifty['chol'],
                     c=over_fifty['target'])

#customize ax0
ax0.set(title='Heart diease and cholestrol levels',
      ylabel='Cholestrol')

# legend ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# adding horizontal line on ax0
ax0.axhline(over_fifty['chol'].mean(),
          linestyle='--')

# adding data to ax1
scatter_1 = ax1.scatter(x=over_fifty['age'],
                     y=over_fifty['thalach'],
                     c=over_fifty['target'])

#customize ax1
ax1.set(title='Heart diease and Max Heart Rate',
       xlabel='Age',
      ylabel='Thalach')

# legend ax1
ax1.legend(*scatter_1.legend_elements(), title='Target')

# adding horizontal line on ax1
ax1.axhline(over_fifty['thalach'].mean(),
          linestyle='--')

# Add title to entire figure
fig.suptitle('Heart Diease Analysis', fontsize=16, fontweight='bold')

plt.show()
```

customizing the plot

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10),
                              nrows=2,
                              ncols=1,
                              sharex=True)

# adding data to ax0
scatter = ax0.scatter(x=over_fifty['age'],
                     y=over_fifty['chol'],
                     c=over_fifty['target'],
                     cmap='winter') #changes color scheme

#customize ax0
ax0.set(title='Heart diease and cholestrol levels',
      ylabel='Cholestrol')

# changing x-axis limits
ax0.set_xlim([50, 80])

# legend ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# adding horizontal line on ax0
ax0.axhline(over_fifty['chol'].mean(),
          linestyle='--')

# adding data to ax1
scatter_1 = ax1.scatter(x=over_fifty['age'],
                     y=over_fifty['thalach'],
                     c=over_fifty['target'],
                     cmap='winter')

#customize ax1
ax1.set(title='Heart diease and Max Heart Rate',
       xlabel='Age',
      ylabel='Thalach')

# changing x-axis limits
ax1.set_xlim([50, 80])
# changing y-axis limits
ax1.set_ylim([60, 200])

# legend ax1
ax1.legend(*scatter_1.legend_elements(), title='Target')

# adding horizontal line on ax1
ax1.axhline(over_fifty['thalach'].mean(),
          linestyle='--')

# Add title to entire figure
fig.suptitle('Heart Diease Analysis', fontsize=16, fontweight='bold')

plt.show()
```
